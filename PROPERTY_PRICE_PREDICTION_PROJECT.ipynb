```python
df = df.drop(['id', 'Date', 'Postal Code',
              'waterfront present','number of views',
              'condition of the house','Area of the house(excluding basement)',
              'Area of the basement','Renovation Year','Lattitude','Longitude',
              'living_area_renov','lot_area_renov'], axis=1)
```

```python
x=df.drop("Price", axis=1)
y=df["Price"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(
    x,y,test_size=0.2,random_state=42
)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
